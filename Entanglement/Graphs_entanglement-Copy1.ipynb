```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools
import numpy as np
```

```python
nlayers = [1, 2, 3, 4, 5]
m = ['s', 'X', 'D','*']
c = ['#98D2EB', '#D57A66', '#FFB7C3', '#4E4B5C']

#Font sizes settings
plt.rc('xtick', labelsize=16)    # fontsize of the tick labels
plt.rc('ytick', labelsize=16)    # fontsize of the tick labels
plt.rc('legend', fontsize=17)    # legend fontsize
plt.rcParams['axes.titley'] = 1.02
```

```python
fig = plt.figure(figsize = (13.0, 8), tight_layout=True)

gs = fig.add_gridspec(2, 2, hspace=0.135, wspace=0.05)
(ax0, ax1), (ax2, ax3) = gs.subplots(sharey='row')

#=================================================================================================================

qubits = 3

entang = {}
error = {}
for layers in nlayers:
    datalocal = pd.read_csv('entanglement_ansatz1_distnormal_qubits{0}_layers{1}_samples10000.txt'
                            .format(qubits,layers))
    entang[layers] = datalocal['MW_mean']
ids = datalocal['Circuit']
    
for k in nlayers:
    entang[k] = [complex(value).real for value in entang[k]]
    
circuits = ids
j=1
for i in itertools.islice(circuits, 1, 6):
    if i == 'Linear':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits-1)*np.array(nlayers)
        ax0.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1
    if i == 'Ring':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits)*np.array(nlayers)
        ax0.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1
    if i == 'Star':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits-1)*np.array(nlayers)
        ax0.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1

mean_MW = (2**qubits - 2)/(2**qubits + 1)
ax0.axhline(mean_MW, color='k', linestyle = 'dashed', linewidth = 1.5)
    

ax0.annotate('(a)', xy=(0.990, 0.89), xycoords='axes fraction', fontsize=22,
                horizontalalignment='right', verticalalignment='bottom')

ax0.xaxis.label.set_size(22)
ax0.yaxis.label.set_size(22)    
    
ax0.set_ylim(0.38, 1.04)
ax0.set_xlim(4,47)
ax0.set_ylabel('Entang. mean')
#ax0.set_xlabel('Number of layers')

#=================================================================================================================

qubits = 4

entang = {}
error = {}
for layers in nlayers:
    datalocal = pd.read_csv('entanglement_ansatz1_distnormal_qubits{0}_layers{1}_samples10000.txt'
                            .format(qubits,layers))
    entang[layers] = datalocal['MW_mean']
ids = datalocal['Circuit']
    
for k in nlayers:
    entang[k] = [complex(value).real for value in entang[k]]
    
circuits = ids
j=1
for i in itertools.islice(circuits, 1, 6):
    if i == 'Linear':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits-1)*np.array(nlayers)
        ax1.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1
    if i == 'Ring':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits)*np.array(nlayers)
        ax1.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1
    if i == 'Star':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits-1)*np.array(nlayers)
        ax1.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1

mean_MW = (2**qubits - 2)/(2**qubits + 1)
ax1.axhline(mean_MW, color='k', linestyle = 'dashed', linewidth = 1.5)
    

ax1.annotate('(b)', xy=(0.990, 0.89), xycoords='axes fraction', fontsize=22,
                horizontalalignment='right', verticalalignment='bottom')

ax1.xaxis.label.set_size(22)
ax1.yaxis.label.set_size(22)    

ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
ax1.set_ylim(0.38, 1.04)
ax1.set_xlim(5,65)
#ax0.set_ylabel('Entanglement mean')
#ax0.set_xlabel('Number of layers')

#=================================================================================================================

qubits = 5

entang = {}
error = {}
for layers in nlayers:
    datalocal = pd.read_csv('entanglement_ansatz1_distnormal_qubits{0}_layers{1}_samples10000.txt'
                            .format(qubits,layers))
    entang[layers] = datalocal['MW_mean']
ids = datalocal['Circuit']
    
for k in nlayers:
    entang[k] = [complex(value).real for value in entang[k]]
    
circuits = ids
j=1
for i in itertools.islice(circuits, 1, 6):
    if i == 'Linear':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits-1)*np.array(nlayers)
        ax2.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1
    if i == 'Ring':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits)*np.array(nlayers)
        ax2.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1
    if i == 'Star':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits-1)*np.array(nlayers)
        ax2.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1

mean_MW = (2**qubits - 2)/(2**qubits + 1)
ax2.axhline(mean_MW, color='k', linestyle = 'dashed', linewidth = 1.5)
    

ax2.annotate('(c)', xy=(0.990, 0.89), xycoords='axes fraction', fontsize=22,
                horizontalalignment='right', verticalalignment='bottom')

ax2.xaxis.label.set_size(22)
ax2.yaxis.label.set_size(22)    

ax2.set_ylim(0.38, 1.04)
ax2.set_xlim(5,80)
ax2.set_ylabel('Entang. mean')
ax2.set_xlabel('Number of gates')

#=================================================================================================================

qubits = 6

entang = {}
error = {}
for layers in nlayers:
    datalocal = pd.read_csv('entanglement_ansatz1_distnormal_qubits{0}_layers{1}_samples10000.txt'
                            .format(qubits,layers))
    entang[layers] = datalocal['MW_mean']
ids = datalocal['Circuit']
    
for k in nlayers:
    entang[k] = [complex(value).real for value in entang[k]]
    
circuits = ids
j=1
for i in itertools.islice(circuits, 1, 6):
    if i == 'Linear':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits-1)*np.array(nlayers)
        ax3.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1
    if i == 'Ring':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits)*np.array(nlayers)
        ax3.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1
    if i == 'Star':
        yentang = []
        for k in nlayers:
            yentang.append(entang[k][j])
        nlayers_local = (3*qubits-1)*np.array(nlayers)
        ax3.plot(nlayers_local, yentang, marker = m[j], color = c[j], markersize=10, linewidth=2.0, label=i)
        j+=1

mean_MW = (2**qubits - 2)/(2**qubits + 1)
ax3.axhline(mean_MW, color='k', linestyle = 'dashed', linewidth = 1.5)
    

ax3.annotate('(d)', xy=(0.990, 0.89), xycoords='axes fraction', fontsize=22,
                horizontalalignment='right', verticalalignment='bottom')

ax3.xaxis.label.set_size(22)
ax3.yaxis.label.set_size(22)    

ax3.set_ylim(0.38, 1.04)
ax3.set_xlim(5,95)
#ax3.set_ylabel('Entang. mean')
ax3.set_xlabel('Number of gates')

fig.savefig('entanglement_normaldist_ngates_ansatz_1_3-4-5-6_qubits.pdf', bbox_inches='tight')
```
